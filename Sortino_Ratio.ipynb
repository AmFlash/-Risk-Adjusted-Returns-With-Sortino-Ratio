# Sortino Ratio 
Sortino-ratio-calculation.jpg.webp

```python
#calculate log returns because they are more accurate with time
log_returns = np.log(1+ closing_price.pct_change()) 

symbols = ['NESTLEIND.NS', 'RELIANCE.NS', 'ADANIPOWER.NS']
no_of_symbols=3
```

```python
Portfolios = int(input("Enter no_of_Portfolios:"))

random_weights = np.zeros((Portfolios,no_of_symbols))

ret_arr = np.zeros(Portfolios)

sortino_arr = np.zeros(Portfolios)

for ind in range(Portfolios):
    
    weights = np.array(np.random.random(no_of_symbols))
    weights  = weights/np.sum(weights)
    
    random_weights[ind, :] = weights
    
    ret_arr[ind] = np.sum(log_returns.mean()*weights*987)
    
    portfolio_return = np.sum(ret_arr[ind]* weights)

    risk_free_rate = 0
    target_return =  2

    downside_deviation = np.sqrt(np.mean(np.maximum(0, target_return - ret_arr[ind]) ** 2))

    sortino_arr[ind] = (portfolio_return - risk_free_rate) / downside_deviation
```

```python
simulation_data = [ret_arr, sortino_arr, random_weights]

simulations_df = pd.DataFrame(data=simulation_data).T

simulations_df.columns = ['Returns','Sortino Ratio','Portfolio weights']

simulations_df = simulations_df.infer_objects()

print('Portfolio metrics:')
display(simulations_df.head())
```
